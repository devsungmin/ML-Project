```python
from sklearn.model_selection import train_test_split
```

```python
# 전체 데이터 불러옴
df=pd.read_csv("./amazon.csv",encoding="ISO-8859-1")
# 전체 데이터 크기 확인
print(df.shape)
```

```python
# 데이터 칼럼값 확인
print(df.columns)

# date데이터의 경우 자세한 날짜가 아닌 매달 1일로 설정되어있어, 파라미터값 삭제
del df['date']
print(df.columns)
```

```python
# Acre지역의 데이터만 불러옴
data = df[0:239]
```

```python
print(data)
print(data.shape)
```

```python
# Acre지역의 월별 산불 횟수
pivot1 = pd.pivot_table(data,values="number",index=["year","month"],aggfunc=np.sum)
```

```python
# Acre지역의 월별 산불 횟수 그래프로 시각화
plt.figure(figsize=(15, 6))
sns.set_style("darkgrid")
ax = sns.barplot(x=pivot1.index, y="number", data=pivot1)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.xlabel("Month and Year")
plt.ylabel("Count of fires")
plt.title("Fire vs Month")
```

```python
# x에 지역, 월
# y는 횟수
X = pivot1.iloc[:,:1]
y = pivot1.iloc[0:,:].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
inertia =[] 
for k in range(1,13):
    km = KMeans(n_clusters=k,algorithm="auto", random_state=0).fit(x_train,y_train)
    
    # 배열에 KMeans fit 값 저장
    inertia.append(km.inertia_)
```

```python
#inertia 배열에 저장된 값을 출력
ax = plt.plot(range(1,13),inertia)
```
